# Tokenisasi & Stemming


```python
import nltk.corpus
nltk.download('stopwords')
from nltk.corpus import stopwords
stop = stopwords.words('indonesian')
from nltk.tokenize import sent_tokenize, word_tokenize
import pandas as pd
```

```python


# Replace 'path/to/your/file.csv' with the actual path to your CSV file
df = pd.read_csv('03_hasil-stopword-filter_processed.csv') 

# Display the first few rows of the DataFrame
df.head() 
```

```python
# Pastikan tidak ada nilai NaN di kolom text_no_stopwords
df['text_no_stopwords'] = df['text_no_stopwords'].fillna('')

# Tokenisasi kolom text_no_stopwords
df['text_tokens'] = df['text_no_stopwords'].apply(lambda x: word_tokenize(x))

# Lihat hasil
df.head(20)
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Inisialisasi stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Fungsi untuk stem kata
def stemmed_wrapper(term):
    return stemmer.stem(term)

# Membuat dictionary term unik dan stemmed
def create_stem_dict(documents):
    """Membuat kamus untuk setiap term yang unik dan melakukan stemming"""
    terms = set(term for document in documents for term in document)  # Ambil term unik
    return {term: stemmed_wrapper(term) for term in terms}

# Membuat kamus stemmed terms dari kolom 'text_tokens'
term_dict = create_stem_dict(df['text_tokens'])

# Fungsi untuk mendapatkan stemmed terms dari dokumen
def get_stemmed_term(document):
    return [term_dict[term] for term in document]

    # Looping untuk memproses setiap dokumen dan kata dalam 'text_tokens'
for document in df['text_tokens']:
    for term in document:
        if term not in term_dict:
            # Jika term belum ada di kamus, tambahkan
            term_dict[term] = stemmed_wrapper(term)
            hitung += 1
            # Cetak proses stemming
            print(f"{hitung}: {term} -> {term_dict[term]}")

print(len(term_dict))
print("------------------------")
for term in term_dict:
    term_dict[term] = stemmed_wrapper(term)
    hitung+=1
    print(hitung,":",term,":" ,term_dict[term])

print(term_dict)
print("------------------------")

# Menerapkan stemming ke dataframe
df['text_steamindo'] = df['text_tokens'].apply(lambda x: ' '.join(get_stemmed_term(x)))

# Tampilkan hasil
df.head(20)
```
