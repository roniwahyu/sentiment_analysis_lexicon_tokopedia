```python
#import library yang dibutuhkan
import pandas as pd
import googletrans
from googletrans import Translator
import numpy as np
```

```python
#import dataset
df = pd.read_csv("/content/tokopedia.csv")
df.head()
```

```python
#SORT DATA, AMBIL YANG DIBUTUHIN AJA
df = df[['text_steamindo']]
df.rename(columns = {"text_steamindo" : "Ulasan"}, inplace=True) #rename
df.head()
```

translate

```python
translator = Translator()
translations = {}
for column in df.columns:
  #unique elements of the column
  unique_elements = df[column].unique()
  for element in unique_elements:
    #adding all the translation to dictionary (translations)
    translations[element] = translator.translate(element).text
translations
```

```python
#mengganti semua kata yang diterjemahkan dari kamus ke original dataframe
df.replace(translations, inplace = True)
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()
```

```python
scores = [analyser.polarity_scores(x) for x in df['Ulasan']]
print(scores)
df['Compound_Score'] = [x['compound'] for x in scores]
```

```python
df.head(10)
```

```python
#compound score lexicon based
df.loc[df['Compound_Score'] < 0, 'Sentiments'] = 'Negatif'

df.loc[df['Compound_Score'] == 0, 'Sentiments'] = 'Netral'

df.loc[df['Compound_Score'] > 0, 'Sentiments'] = 'Positif'
df.head(10)
```
